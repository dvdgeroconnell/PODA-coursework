```python
# Data frames
import pandas as pd

# Plotting - import the main plotting functionality we need  
import matplotlib.pyplot as plt

# Import numpy
import numpy as np
```

## Load Data

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
```

```python
# Summary of data.
df.describe()
```

## Two Variable Plots  
***

```python
# Get just the petal lengths
plen = df['petal_length']
# Each column of a data frame is a data series - plen is the latter
print(type(plen))
# Get the numpy array - more basic than pandas data frames (spreadsheet) or data series (column)
plen.to_numpy()
```

```python
pwidth = df['petal_width'].to_numpy()
pwidth
```

```python
# Plot plen and pwidth against each other - simple plot to start with
plt.plot(plen, pwidth, 'x')
#plt.plot(plen, pwidth, 'x', scalex=0.5, scaley = 0.5)
# Axis labels and Title
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Data Set')

# Change the x-axis limits
plt.xlim(0,8)

# Change the y-axis limits
plt.ylim(0,4)

# There's no need for plt.show() because Jupyter is set up to do that already.

# Note this is a stateful approach as the library is remembering things between calls to the plot function
# That's why we can follow up the plot command with axis, label and title function calls...
```

```python
# Use plt.subplots() for stateless
# See https://realpython.com/python-matplotlib-guide/#understanding-pltsubplots-notation

# Default is one axis - this does 3 rows, 2 columns
# fig, ax = plt.subplots(3,2)

fig, ax = plt.subplots()

# Fill in the axes. If your have more than one, you need to specify which one.
ax.plot(plen, pwidth, 'x')

# Can do these before or after the plot.
ax.set_xlabel('Petal Length (cm)')
ax.set_ylabel('Petal Width (cm)')
ax.set_title('Iris Data Set')
ax.set_xlim(0,8)
ax.set_ylim(0,4)
```

## Add a Best Fit Line  
***  


$  y = mx + c = p_1 x^1 + p_0 = p_1 x + p_0 $

p1 and p0 are what numpy is looking for when it does a polyfit.

```python
# Fit a straight line between x & y
# See - https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html

m, c = np.polyfit(plen, pwidth, 1)

# Show m and c
m, c
```

```python
# x values for best fit line
# This gives you an array between 0 and 8 (both included) and 100 values
# Could use plen.max()+1.0 instead of 8.0 to make it a bit more general...
bf_x = np.linspace(0.0,8.0,100)

# y values for best fit line
bf_y = m * bf_x + c

# Show them
# bf_x, bf_y
```

```python

# Draw the simple plot, specify the axis scales and titles
fig, ax = plt.subplots()
ax.plot(plen, pwidth, 'x')

# Now add a second plot with the best fit line on top of the first
# 'r-' means colour the line red and draw it as a line
ax.plot(plen, m * plen + c, 'r-')

ax.set_xlabel('Petal Length (cm)')
ax.set_ylabel('Petal Width (cm)')
ax.set_title('Iris Data Set')
ax.set_xlim(0,8)
ax.set_ylim(0,4)
```

```python
# x values for best fit line
# This gives you an array between 0 and 8 (both included) and 100 values
# Could use plen.max()+1.0 instead of 8.0 to make it a bit more general...
bf_x = np.linspace(0.0,8.0,100)

# y values for best fit line
bf_y = m * bf_x + c
```

```python
# Draw the simple plot, specify the axis scales and titles
fig, ax = plt.subplots()
ax.plot(plen, pwidth, 'x')

# Now add a second plot on top of the first with the best fit line
# This time, use the values from the arrays 
# 'r-' means colour the line red and draw it as a line
ax.plot(bf_x, bf_y, 'r-')

ax.set_xlabel('Petal Length (cm)')
ax.set_ylabel('Petal Width (cm)')
ax.set_title('Iris Data Set')
ax.set_xlim(0,8)
ax.set_ylim(-1,4)

# Suggests that for very small petal lengths, the petal widths will be negative
# So it should really go through the origin - there's some way to force this (c should be zero)
# But the fit would be less accurate in this case
# It's extrapolation below the lowest real data point (and beyond the highest point) anyway
# So don't go there...
```

### Pearson correlation coefficient (r)  

```python
# Measure the correlation
# See https://numpy.org/doc/stable/reference/generated/numpy.corrcoef.html

np.corrcoef(plen, pwidth)

# Note in the returned array the '1' in the top left implies that plen is perfectly correlated
# with itself and in the bottom right implies that pwidth is perfectly correlated with itself.
# Top right is plen correlation with pwidth - above 0.8 is strong correlation.
# Bottom left is the pwidth correlation with plen - same result, as would be expected.
# See https://en.wikipedia.org/wiki/Pearson_correlation_coefficient
# And - https://statistics.laerd.com/statistical-guides/pearson-correlation-coefficient-statistical-guide.php

```

## Further Plots and Correlation  
***  

### Sepal Plots and Correlation  

```python
# Get the numpy arrays for sepal length and width
slen = df['sepal_length'].to_numpy()
swidth = df['sepal_width'].to_numpy()
```

```python
# Calculate a best fit straight line
m1, c1 = np.polyfit(slen, swidth, 1)

# Show m1 and c1
m1, c1
```

```python
# Stateless for sepal width vs length

fig1, ax1 = plt.subplots()

# Fill in the axes. We have just one.
ax1.plot(slen, swidth, 'x')

# Now add a second plot with the best fit line on top of the first
# 'r-' means colour the line red and draw it as a line
ax1.plot(slen, m1 * slen + c1, 'r-')

# Can do these before or after the plot.
ax1.set_xlabel('Sepal Length (cm)')
ax1.set_ylabel('Sepal Width (cm)')
ax1.set_title('Iris Data Set')
ax1.set_xlim(0,8)
ax1.set_ylim(0,5)
```

```python
# Calculate the correlation - it looks like it will be a lot lower than for petal length vs width
np.corrcoef(slen, swidth)
```

### Combine Plots as Multiple Axes in a Figure
xx 

```python

# Default is one axis - this does 3 rows, 2 columns
fig2, (ax3, ax4) = plt.subplots(1,2)

# Fill in the axes. If your have more than one, you need to specify which one.
ax3.plot(plen, pwidth, 'x')
ax4.plot(slen, swidth, 'x')

# Set the axis limits and titles for the Petal axis
ax3.set_xlabel('Petal Length (cm)')
ax3.set_ylabel('Petal Width (cm)')
ax3.set_title('Iris Petal dataset')
ax3.set_xlim(0,8)
ax3.set_ylim(0,3)

# Set the axis limits and titles for the Sepal axis
ax4.set_xlabel('Sepal Length (cm)')
ax4.set_ylabel('Sepal Width (cm)')
ax4.set_title('Iris Sepal dataset')
ax4.set_xlim(0,8)
ax4.set_ylim(0,5)

# Set the title for the figure
fig2.suptitle('Iris Data Set')

# Overlay the best fit lines calculated in an earlier cell
ax3.plot(plen, m * plen + c, 'r-')
ax4.plot(slen, m1 * slen + c1, 'r-')
```
